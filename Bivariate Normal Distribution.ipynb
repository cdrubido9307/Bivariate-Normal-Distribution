# Bivariate Normal distribution

$$ f(x, y) = \frac{1}{2 \pi \sigma_X \sigma_Y \sqrt{1 - \rho^2}} e^{-\frac{1}{2 \left(1 - \rho^2\right)} \left( \left( \frac{x - \mu_X}{\sigma_X} \right)^{2} + \left(\frac{y - \mu_Y}{\sigma_Y}\right) ^{2} - 2 \rho \frac{\left(x - \mu_X\right)\left(y - \mu_Y\right)}{\sigma_X \sigma_Y}\right)} $$

Consider,
$$ \mu_X = 0 $$
$$ \mu_Y = 0 $$
$$ \sigma_X^{2} = 1 $$
$$ \sigma_Y^{2} = 1 $$
$$ \rho = \frac{1}{2} $$

Then we have,
$$ f(x, y) = \frac{1}{2 \pi \sqrt{1 - \left( \frac{1}{2} \right) ^ 2}} e^{-\frac{1}{2\left(1 - \left(\frac{1}{2}\right)^2\right)} \left(x^{2} + y^{2} - xy \right)} $$

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return (1/(2*np.pi * np.sqrt(1 - 0.25))) * np.exp((-1/(2* (1 - 0.25)))*(x**2 + y**2 - x*y))
```

```python
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('Bivariate Normal distribution');
ax.view_init(60, 35)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Bivariate Normal distribution');
ax.view_init(60, 35)
```

Using the same parameters but changing $ \rho = 0.8 $,

```python
def g(x, y):
    return (1/2*np.pi * np.sqrt(1 - (0.8)**2)) * np.exp(-1/(2*(1-0.8))*(x**2 + y**2 - (2 * 0.8) * x*y))
```

```python
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z1 = g(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z1, color='black')
ax.set_title('Bivariate Normal distribution');
ax.view_init(60, 35)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z1, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('Bivariate Normal distribution');
ax.view_init(60, 35)
```
